```
import seaborn as sns
import pandas as pd
```

```
passageiros = pd.read_csv('https://iaexpert.academy/wp-content/uploads/2019/10/AirPassengers.csv')
passageiros
# este dataset é utilizado geralmente para séries temporais
```

Gráfico de linha

```
sns.relplot(x='Month',y='#Passengers',kind='line',height=10,data=passageiros);
# conforme o tempo passa a quantidade de passageiros aumenta
```
